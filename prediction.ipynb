```python
import pandas as pd
```

```python
# confusion matrix
# dla 100 związków

tp = 5.0
tn = 80.0
fp = 10.0
fn = 5.0
p = 10.0
n = 90.0
```

```python
#Sensitivity - 'czulosc testu' - ile prawdziwie pozytywnych zostalo znalezionych
Se = tp /(tp+fn)
#Specificity - Swoistość testu - ile prawdziwie negatywnych zostało znalezionych
Sp = tn /(tn+fp)
#Precision - 'Precyzja testu' -ile z pozytywnych predykcji jest rzeczywiście prawdziwych
Ppv = tp /(tp+fp)
#Accuracy - 'Dokładność testu' - ile predykcji okazało się prawidłowych
Acc = (tp + tn)/(p + n)
```

```python
print("Se = %.2f" % Se)
print("Sp = %.2f" % Sp)
print("Ppv = %.2f" % Ppv)
print("Acc = %.2f" % Acc)
```

```python
data_actives = pd.read_csv('actives_docked.csv') #czasteczki aktywne
data_decoys = pd.read_csv('decoys_docked.csv') #czasteczki nieaktywne
```

```python
data_actives.head()
```

```python
data_decoys.head()
```

```python
data_decoys['act_bin'] = 0
data_actives['act_bin'] = 1
```

```python
#CONCATANATE

data = pd.concat([data_actives, data_decoys])
```

```python
data.head()
```

```python
data = data[['name', 'score', 'act_bin']]
```

```python
data = data.drop_duplicates('name')
data = data.sort_values('score', ascending=True)
```

```python
data.head()
```

```python
predict = (-data['score']).tolist()
experim = (data['act_bin']).tolist()
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
```

```python
fpr, tpr, thresh = roc_curve(experim, predict)
roc_auc = auc(fpr,tpr) 
```

```python
plt.title("ROC")
plt.plot(fpr,tpr, 'b', label="AUC = %.2f" % roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1],[0,1], 'r--')
plt.xlabel("FPR")
plt.ylabel("TPR")

plt.show()
```
